```python
import json
import time
import pandas as pd
from kafka import KafkaProducer

# Load dataset
df = pd.read_csv("IoT_Intrusion.csv")

# Kafka producer
producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v, default=str).encode("utf-8")
)

topic_name = "iot-topic"

batch_size = 50   # how many records per batch (you can change this)

# Loop in batches
for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i+batch_size]

    print(f"\n🚀 Sending batch {i//batch_size + 1}...")

    for _, row in batch.iterrows():
        message = row.to_dict()
        producer.send(topic_name, value=message)

    producer.flush()  # send all messages in batch

    print(f"✅ Sent {len(batch)} records. Waiting 15 seconds...\n")
    time.sleep(15)

producer.close()
print("🎉 Done sending all data!")
```
